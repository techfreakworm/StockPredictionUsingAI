```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
```

```python
ex = 'European authorities fined Google a record $5.1 billion on Wednesday for abusing its power in the mobile phone market and ordered the company to alter its practices'
```

```python
def preprocess(sent):
    sent = nltk.word_tokenize(sent)
    sent = nltk.pos_tag(sent)
    return sent
```

```python
sent = preprocess(ex)
sent
```

```python
pattern = 'NP: {<DT>?<JJ>*<NN>}'
```

```python
cp = nltk.RegexpParser(pattern)
cs = cp.parse(sent)
print(cs)
```

```python
from nltk.chunk import conlltags2tree, tree2conlltags
from pprint import pprint
iob_tagged = tree2conlltags(cs)
pprint(iob_tagged)
```

```python
ne_tree = nltk.ne_chunk(pos_tag(word_tokenize(ex)))
print(ne_tree)
```

## Extracting named entity from an article


```python
from bs4 import BeautifulSoup
import requests
import re
def url_to_string(url):
    res = requests.get(url)
    html = res.text
    soup = BeautifulSoup(html, 'html.parser')
    for script in soup(["script", "style", 'aside']):
        script.extract()
    return " ".join(re.split(r'[\n\t]+', soup.get_text()))
ny_bb = url_to_string('https://www.nytimes.com/2018/08/13/us/politics/peter-strzok-fired-fbi.html?hp&action=click&pgtype=Homepage&clickSource=story-heading&module=first-column-region&region=top-news&WT.nav=top-news')
article = nlp(ny_bb)
len(article.ents)
```

```python
labels = [x.label_ for x in article.ents]
Counter(labels)
```

```python
items = [x.text for x in article.ents]
Counter(items).most_common(3)
```

```python
sentences = [x for x in article.sents]
print(sentences[21])
```

```python
displacy.render(nlp(str(sentences[21])), style='dep', jupyter = True, options = {'distance': 120})
```

```python
[(x.orth_,x.pos_, x.lemma_) for x in [y 
                                      for y
                                      in nlp(str(sentences[21])) 
                                      if not y.is_stop and y.pos_ != 'PUNCT']]
```

```python
dict([(str(x), x.label_) for x in nlp(str(sentences[21])).ents])
```

```python
print([(x, x.ent_iob_, x.ent_type_) for x in sentences[21]])
```

```python
displacy.render(nlp(''.join([str(x) for x in sentences])), jupyter=True, style='ent')
```
